```python
aapl_df = df[df['Name']=="AAPL"] 
amzn_df = df[df['Name']=="AMZN"] 
fb_df = df[df['Name']=="FB"] 
googl_df = df[df['Name']=="GOOGL"] 
msft_df = df[df['Name']=="MSFT"] 
```

```python
plt.figure(figsize=(25,25))
plt.plot('date', 'close', data = aapl_df, label='Apple')
plt.plot('date', 'close', data = amzn_df, label='Amazon')
plt.plot('date', 'close', data = fb_df, label='Facebook')
plt.plot('date', 'close', data = googl_df, label='Google')
plt.plot('date', 'close', data = msft_df, label='Microsoft')
plt.xticks(np.arange(0, 1250, 120))
plt.legend()
plt.show()
```
